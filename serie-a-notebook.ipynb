```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("serie_a_cleaned.xlsx")
```

```python
df.head()
```

```python
df.head()
```

```python
plt.figure(figsize=(50,35))
plt.title('Points / Market value')
sns.set(style='whitegrid', context='talk', font_scale=3)
ax = sns.scatterplot(x='MV',y='Points',data=df)


for line in range(0,df.shape[0]):
    ax.text(df.MV[line], df.Points[line], df.Team[line],  horizontalalignment='center', size='20', color='black', weight='semibold')

ax;
```

```python
plt.figure(figsize=(50,35))
# plt.title('Points / Market value in 2019-20 Serie A season')
# plt.style.use('dark_background')
sns.set(style='whitegrid', context='talk', font_scale=3)
ax2 = sns.scatterplot(x='MV',y='Points',data=df)
plt.xlabel("\n Market value in millions of ")
plt.ylabel("Points \n")
plt.ylim(0,90)
plt.xlim(0,800)
ax2;
```

```python
plt.figure(figsize=(50,35))
# plt.title('Points / Market value in 2019-20 Serie A season')
# plt.style.use('dark_background')
sns.set(style='whitegrid', context='talk', font_scale=3)
ax2 = sns.regplot(x='MV',y='Points',data=df, ci=None)
plt.xlabel("\n Market value in millions of ")
plt.ylabel("Points \n")
plt.ylim(0,90)
plt.xlim(0,800)
ax2;
```
